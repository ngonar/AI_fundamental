```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
irisMLP = MLPClassifier(hidden_layer_sizes=(3), learning_rate_init=0.1, momentum=0.9, max_iter=1000)

iris_full = datasets.load_iris()
iris_input = iris_full['data']
iris_target = iris_full.target

input_train, input_test, target_train, target_test = train_test_split(iris_input, iris_target, test_size=0.3)

scaler = StandardScaler()
scaler.fit(iris_input)

input_train_std = scaler.transform(input_train)
input_test_std = scaler.transform(input_test)
```

```python
iris_input
```

```python
irisMLPTrained = irisMLP.fit(input_train_std, target_train)
```

```python
irisMLPPredict = irisMLP.predict(input_test_std)
```

```python
from sklearn.metrics import accuracy_score, cohen_kappa_score

pct_accuracy = accuracy_score(target_test, irisMLPPredict)

print('%.2f%%' % (pct_accuracy*100))
```

```python
from sklearn.metrics import cohen_kappa_score

kappa = cohen_kappa_score(target_test, irisMLPPredict)

print('%.2f'%kappa)
```

```python
import matplotlib.pyplot as plt
X=range(0,len(target_test))
plt.figure()
plt.plot(X,target_test, color='blue', label='Target', linewidth=2)
plt.plot(X,irisMLPPredict, color='red', label='Predicted', linewidth=2)
plt.xlabel('Example')
plt.ylabel('Iris')
plt.title('MLP Perceptron - Iris')
plt.legend()
plt.show()
```
