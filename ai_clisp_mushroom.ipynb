```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\admin\\Downloads\\mushroom_with_featurename.csv")
print(len(data))
data.head()
```

```python
mushroomIn = data[["feature_7","feature_21","feature_22","feature_27","feature_37","feature_52","feature_63","feature_67","feature_97","feature_104","feature_110"]]
```

```python
mushroomIn.shape
mushroomIn.head()

```

```python
mushroomOut = data['FLAG']
mushroomOut.shape
mushroomOut.head()
```

```python
isMushroom = ['true','false']
variablesList = ["feature_7","feature_21","feature_22","feature_27","feature_37","feature_52","feature_63","feature_67","feature_97","feature_104","feature_110"]

    
print(len(variablesList))
```

```python
# for index, row in data.iterrows():
#     print(row)
```

```python
outputValues = []
for index,currRow in mushroomIn.iterrows():
    #print(currRow)
    for i in range(0,len(variablesList)):
        #print(currRow[i])
        sFact = "("
        fValue = currRow[i]
        sValue = str(fValue)
        sName = variablesList[i]
        sFact = sFact + sName + ' ' + sValue + ")"
        #print(sFact)
        fact = env.assert_string(sFact)
        
    r = env.run()
    
    facts = env.facts()
    iOutputValue = 0
    for f in facts:
        sFactName = f.template.name
        #print(f, sFactName)

        if sFactName == 'Mushroom':
            sValue = f[0]
            if sValue == "true":
                iOutputValue = 1

    outputValues.append(iOutputValue)
    
    env.reset()
```

```python
outputValues
```

```python
from sklearn.metrics import accuracy_score
fPct = accuracy_score(mushroomOut, outputValues)
fPct = fPct * 100
print('Percent: %.2f'%fPct)
```

```python
from sklearn.metrics import cohen_kappa_score
fKappa = cohen_kappa_score(mushroomOut, outputValues)
print('Kappa: %.2f'%fKappa)
```
