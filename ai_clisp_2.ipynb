```python
import clips
env = clips.Environment()
```

```python
env.load("C:\\Users\\admin\\Downloads\\rules_clisp.clp")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
irisIn, irisOut = iris.data, iris.target

irisIn.shape
```

```python
irisOut.shape
print(irisOut)
```

```python
speciesList = ['Iris Setosa', 'Iris Versicolor', 'Iris Virginica']
variablesList = ['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth']
```

```python
outputValues = []
for currRow in irisIn:
    print(currRow)
    for i in range(0,4):
        sFact = "("
        fValue = currRow[i]
        sValue = str(fValue)
        sName = variablesList[i]
        sFact = sFact + sName + ' ' + sValue + ")"
        print(sFact)
        fact = env.assert_string(sFact)
        
    r = env.run()
    
    facts = env.facts()
    for f in facts:
        sFactName = f.template.name
        if sFactName == 'Species':
            sValue = f[0]
            iOutputValue = speciesList.index(sValue)
            
    outputValues.append(iOutputValue)
    
    env.reset()
```

```python
from sklearn.metrics import accuracy_score
fPct = accuracy_score(irisOut, outputValues)
fPct = fPct * 100
print('Percent: %.2f'%fPct)
```

```python
from sklearn.metrics import cohen_kappa_score
fKappa = cohen_kappa_score(irisOut, outputValues)
print('Kappa: %.2f'%fKappa)
```
