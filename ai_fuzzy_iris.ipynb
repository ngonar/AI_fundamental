```python
import numpy as np
import skfuzzy as fuzz
```

```python
sepal_length_min = 4.3
sepal_length_max = 7.9
sepal_length_range = np.arange(sepal_length_min,sepal_length_max,0.1)

sepal_length_avg = np.mean(sepal_length_range)
sepal_length_sigma = np.std(sepal_length_range)
sepal_length_short = fuzz.zmf(sepal_length_range,sepal_length_min, sepal_length_avg)
sepal_length_medium = fuzz.gaussmf(sepal_length_range, sepal_length_avg, sepal_length_sigma)
sepal_length_long = fuzz.smf(sepal_length_range, sepal_length_avg, sepal_length_max)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, figsize=(11,10))
ax.plot(sepal_length_range, sepal_length_short, 'b', linewidth=1.5, label='Short')
ax.plot(sepal_length_range, sepal_length_medium, 'g', linewidth=1.5, label='Medium')
ax.plot(sepal_length_range, sepal_length_long, 'r', linewidth=1.5, label='Long')
ax.set_title("Sepal Length")
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.tight_layout()
plt.show()
```

```python
species_range = np.arange(0,3,1)
species_setosa = fuzz.trimf(species_range, [0,0,0])
species_versicolor = fuzz.trimf(species_range, [1,1,1])
species_virginica = fuzz.trimf(species_range, [2,2,2])
```

```python
from skfuzzy import control as ctrl

sepal_length = ctrl.Antecedent(sepal_length_range, 'sepal_length')

sepal_length['Short'] = sepal_length_short
sepal_length['Medium'] = sepal_length_medium
sepal_length['Long'] = sepal_length_long
```

```python
sepal_width = ctrl.Antecedent(sepal_width_range, 'sepal_width')

sepal_width['Narrow'] = sepal_width_short
sepal_width['Medium'] = sepal_width_medium
sepal_width['Wide'] = sepal_width_long
```

```python
petal_length = ctrl.Antecedent(petal_length_range, 'petal_length')

petal_length['Short'] = petal_length_short
petal_length['Medium'] = petal_length_medium
petal_length['Long'] = petal_length_long
```

```python
petal_width = ctrl.Antecedent(petal_width_range, 'petal_width')

petal_width['Narrow'] = petal_width_short
petal_width['Medium'] = petal_width_medium
petal_width['Wide'] = petal_width_long
```

```python
species  = ctrl.Consequent(species_range, 'species')
species['Setosa'] = species_setosa
species['Versicolor'] = species_versicolor
species['Virginica'] = species_virginica
```

```python
# 1.	if Sepal Length is Short and Sepal Width is Medium and Petal Length is Short and Petal Width is Narrow then Iris is Iris Setosa
rule1 = ctrl.Rule(sepal_length['Short'] | sepal_width['Medium'] | petal_length['Short'] | petal_width['Narrow'], species['Setosa'])
```

```python
# 2.	if Sepal Length is Short and Sepal Width is Wide and Petal Length is Short and Petal Width is Narrow then Iris is Iris Setosa
rule2 = ctrl.Rule(sepal_length['Short'] | sepal_width['Wide'] | petal_length['Short'] | petal_width['Narrow'], species['Setosa'])
```

```python
# 3.    if Sepal Length is Short and Sepal Width is Narrow and Petal Length is Short and Petal Width is Narrow then Iris is Iris Setosa
rule3 = ctrl.Rule(sepal_length['Short'] | sepal_width['Narrow'] | petal_length['Short'] | petal_width['Narrow'], species['Setosa'])
```

```python
# 4.    if Sepal Length is Medium and Sepal Width is Medium and Petal Length is Short and Petal Width is Narrow then Iris is Iris Setosa
rule4 = ctrl.Rule(sepal_length['Medium'] | sepal_width['Medium'] | petal_length['Short'] | petal_width['Narrow'], species['Setosa']) 
```

```python
# 5.	if Sepal Length is Medium and Sepal Width is Narrow and Petal Length is Medium and Petal Width is Medium then Iris is Iris Versicolour
rule5 = ctrl.Rule(sepal_length['Medium'] | sepal_width['Narrow'] | petal_length['Medium'] | petal_width['Medium'], species['Versicolor'])
```

```python
# 6.	if Sepal Length is Short and Sepal Width is Narrow and Petal Length is Medium and Petal Width is Medium then Iris is Iris Versicolour
rule6 = ctrl.Rule(sepal_length['Short'] | sepal_width['Narrow'] | petal_length['Medium'] | petal_width['Medium'], species['Versicolor'])
```

```python
# 7.	if Sepal Length is Short and Sepal Width is Narrow and Petal Length is Medium and Petal Width is Narrow then Iris is Iris Versicolour
rule7 = ctrl.Rule(sepal_length['Short'] | sepal_width['Narrow'] | petal_length['Medium'] | petal_width['Narrow'], species['Versicolor'])
```

```python
# 8.	if Sepal Length is Medium and Sepal Width is Narrow and Petal Length is Medium and Petal Width is Narrow then Iris is Iris Versicolour
rule8 = ctrl.Rule(sepal_length['Medium'] | sepal_width['Narrow'] | petal_length['Medium'] | petal_width['Narrow'], species['Versicolor'])
```

```python
# 9.	if Sepal Length is Medium and Sepal Width is Medium and Petal Length is Long and Petal Width is Wide then Iris is Iris Virginica
rule9 = ctrl.Rule(sepal_length['Medium'] | sepal_width['Medium'] | petal_length['Long'] | petal_width['Wide'], species['Virginica'])
```

```python
#10.	if Sepal Length is Long and Sepal Width is Medium and Petal Length is Long and Petal Width is Wide then Iris is Iris Virginica
rule10 = ctrl.Rule(sepal_length['Long'] | sepal_width['Medium'] | petal_length['Long'] | petal_width['Wide'], species['Virginica'])
```

```python
#11.	if Sepal Length is Long and Sepal Width is Narrow and Petal Length is Long and Petal Width is Wide then Iris is Iris Virginica
rule11 = ctrl.Rule(sepal_length['Long'] | sepal_width['Narrow'] | petal_length['Long'] | petal_width['Wide'], species['Virginica'])
```

```python
#12.	if Sepal Length is Medium and Sepal Width is Narrow and Petal Length is Long and Petal Width is Wide then Iris is Iris Virginica
rule12 = ctrl.Rule(sepal_length['Medium'] | sepal_width['Narrow'] | petal_length['Long'] | petal_width['Wide'], species['Virginica'])
```

```python
rule_set = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
```

```python
the_species = ctrl.ControlSystemSimulation(rule_set)
```

```python
the_species.input['sepal_length'] = 4.3
the_species.input['sepal_width'] = 3
the_species.input['petal_length'] = 4
the_species.input['petal_width'] = 0.5

the_species.compute()

print(the_species.output)
```
